<h1 style="background-color: #FAF9F6; font-family: newtimroman; font-size: 250%; color: darkblue; border: 12px solid #A9A9A9; border-radius: 20px; padding: 15px; text-align: center;">
    <b>Decision Tree</b>
</h1>

<p style="font-family: Arial, sans-serif; font-size: 110%; color: black; padding: 10px; text-align: justify;">
    Decision Tree, hem classification hem de regresyon problemlerinde kullanılan popüler ve sezgisel bir machine learning algoritmasıdır. Decision Tree, veriyi dallara ayırarak sınıflandırma veya tahmin yapar ve sonuçları leaf node'larda verir. Her bir node, veriyi en iyi şekilde ayıracak bir özelliği ve bu özellik için en iyi bölünme noktasını seçer. Bu süreç, veriyi daha homojen alt gruplara ayırarak devam eder.
</p>


<h2 style="font-family: newtimroman; font-size: 150%; color: darkblue; padding: 10px; text-align: left;">
    Root Node:
</h2>

<p style="font-family: Arial, sans-serif; font-size: 110%; color: black; padding: 10px; text-align: justify;">
    Karar ağacının en üstünde yer alır ve tüm veri setini temsil eder. Ağacın ilk bölünme noktası buradan başlar. Kök düğüm, belirli bir özelliği seçerek veriyi iki veya daha fazla gruba ayırır.
</p>
