# Polynomial Regression - Overfitting Simulation

***Polynomial Regression***, doğrusal olmayan ilişkileri modellemek için kullanılan bir regresyon yöntemidir.

***Overfitting***: 

- Model, eğitim verilerine çok iyi uyum sağlar ancak yeni verilere genelleme yapamaz. Eğitim verisinin gürültüsünü ve ayrıntılarını aşırı öğrenir.

- **Train Score'u Test Score'undan Çok daha yüsek olur.**

- **Train Score'u çok daha yüksek olur**. Model, eğitim verilerindeki detayları ve gürültüyü aşırı derecede öğrenmiş olduğu için eğitim veri setinde çok iyi performans gösterir.

- **Test Score'u çok daha düşük olur**.  Model, eğitim veri setinde öğrendiği özel detayları ve gürültüyü genelleştiremediği için, görmediği veriler üzerinde kötü performans gösterir.

- ***Overfitting durumunu anlamanın anahtar yolu, eğitim skorunun test skorundan önemli ölçüde daha yüksek olmasıdır. Bu, modelin eğitim verilerine aşırı uyum sağladığını ve genelleme yapamadığını gösterir.***


***Underfitting***: 

- Model, eğitim verilerindeki temel yapıyı öğrenmek için yetersiz kalır, bu yüzden hem eğitim hem de test verilerinde kötü performans sergiler. Model çok basittir veya yeterli öğrenme kapasitesine sahip değildir.

- **Train Score'u ve Test Score'u düşük olur.** 

- **Train Score'u Düşük Olur**. Model eğitim verilerindeki ilişkileri yeterince öğrenemediği için eğitim veri setindeki performansı zayıftır.

- **Test Score'u Düşük Olur**. ve eğitim skoruna benzer ya da biraz daha kötü. Model, görmediği veriler üzerinde de benzer şekilde kötü performans gösterir çünkü model eğitim verisindeki örüntüleri bile öğrenemediği için genelleme yapamaz.

- Underfitting durumunda, eğitim ve test skorlarının ikisi de düşük olur; ancak, genellikle eğitim skoru test skorundan biraz daha yüksek olur veya ikisi çok yakın olabilir. 

***Underfitting ile overfitting arasındaki temel fark***, overfitting'de modelin eğitim verilerine aşırı uyum sağlaması ve eğitim skorunun çok yüksek olması, test skorunun ise eğitim skorundan önemli ölçüde düşük olmasıdır. Underfitting'de ise her iki skor da düşüktür ve aralarında büyük bir fark olmayabilir.

**Types of Polinomials**

- 1st degree -> Linear    : b1X + b0

- 2nd degree -> Quadratic : b2X ** 2 + b1X + b0

- 3rd degree -> Cubic     : b3X ** 3 + b2X ** 2 + b1X + b0

# Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Örnek data seti üzerinden değerlendirelim

data = np.array([[2,3,4]])
print(data)
```

```python
# Data'ya 2nd degree bir polynomial features ekleyelim

trans = PolynomialFeatures(degree=2, include_bias=False)
```

```python
trans.fit(data)
```

## NOT!

- **Train datası için**: Önce **".fit"** sonra **".transform"** veya doğruda **".fit_transform"** uygulanır**

- **Test datası için**: Sadece **".transform"** uygulanır

```python
trans.transform(data)

# 2, 3, 4 , 2x3, 2x4, 3x4, 2**2, 3**2, 4**2 bu şekilde 2.dereceye çevirdi transform işlemi tamamladı.
# X1 ** 2 + X2 ** 2 + X3 ** 2 + X1X2 + X1X3 + X2X3 + X1 + X2 + X3 
```

```python
# hem transformer hem de fit işlemini aynı anda yapmak için alttaki kodu da kullanıypruz
# yukarıdaki iki işlemi tek metodla yapılabilyor 

trans.fit_transform(data)
```

degree 3 yapınca hem feature sayısı artıyor hemde işlemin yorumlanması zorlaşıyor ve işler karmaşıklaşıyor (19 deger uretildi)

2, 3, 4, 2x3, 2x4, 3x4, 2**2, 3**2, 4**2, 2x3x4, 3x2**2, 4x2**2, 2x3**2, 4x3**2, 2x4**2, 3x4**2, 2**3, 3**3, 4**3

## Polynomial Converter

**fit():** polinom regresyon modelini veri kümesine uydurmak için kullanılır. Yani, veri kümesindeki örnekleri ve bunların hedef değerlerini alır ve bunlar arasında bir regresyon modeli oluşturur. Model, veriye en iyi uyan katsayıları (ve eğer varsa diğer parametreleri) hesaplar. Bu işlem sırasında, model eğitilir ve veriye uyum sağlar.

**transform():** veriyi dönüştürmek için kullanılır. Özellikle, polinom regresyonunda, veri kümesindeki feature'ların kuvvetlerini alır ve bu kuvvetlerin kombinasyonunu kullanarak yeni feature'lar oluşturur. Bu, polinom regresyon modelini kullanmak için gerekli olan feature'ları oluşturur. Yani, veri kümesini polinom regresyon modeline uygun hale getirir.

**Özetle:**  **.fit()** fonksiyonu modeli eğitmek ve veriye uyum sağlamak için kullanılırken, **.transform()** fonksiyonu veriyi dönüştürmek ve polinom regresyon modeliyle uyumlu hale getirmek için kullanılır.

# Evaluation on the Test Set

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pred = model_poly.predict(X_test) #X_test ile testimi yapıyorum
```

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance:")
    print("--------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
# Test Scores

eval_metric(y_test, y_pred) 
```

```python
y_train_pred = model_poly.predict(X_train)
```

```python
# Train Scores

eval_metric(y_train, y_train_pred)
```

# Find an Optimum Polynomial Degree

**RMSE değerine göre optimum dereceyi bulacağız**

```python
def poly(d):   # degree derecesi içine ne yazarsam ona göre çözüm oluşturacak
    
    train_rmse_errors = []   # trains rmse değerlerin toplayacak
    test_rmse_errors = []    # test rmse değerlerin toplayacak
    number_of_features = []  # derece arttıkça feature sayısının kaça çıktığnı göreceğiz
    
    for i in range(1, d):  # 1 den kaçıncı dereceye kadar istiyorsak ona göre yazacağız
        polynomial_converter = PolynomialFeatures(degree = i, include_bias =False)
        poly_features = polynomial_converter.fit_transform(X)
        
        X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
        
        model = LinearRegression(fit_intercept=True) 
        model.fit(X_train, y_train)
        
        train_pred = model.predict(X_train)
        test_pred = model.predict(X_test)
        
        train_RMSE = np.sqrt(mean_squared_error(y_train,train_pred))
        test_RMSE = np.sqrt(mean_squared_error(y_test,test_pred))
        
        train_rmse_errors.append(train_RMSE)
        test_rmse_errors.append(test_RMSE)
        
        number_of_features.append(poly_features.shape[1])
        
    return pd.DataFrame({"train_rmse_errors": train_rmse_errors, "test_rmse_errors":test_rmse_errors, "number of features":number_of_features}, 
                        index=range(1,d))
```

```python
poly(10)
```

```python
# yukardaki sonucların grafigini çizelim ve patlama noktasını görelim

plt.plot(range(1,10), poly(10)["train_rmse_errors"], label = "TRAIN")
plt.plot(range(1,10), poly(10)["test_rmse_errors"], label = "TEST")
plt.xlabel("Polynamial Complex")
plt.ylabel("RMSE")
plt.legend();
plt.show()
```

```python
# grafige daha yakından bakalım

plt.plot(range(1,6), poly(6)["train_rmse_errors"], label = "TRAIN")
plt.plot(range(1,6), poly(6)["test_rmse_errors"], label = "TEST")
plt.xlabel("Polynamial Complex")
plt.ylabel("RMSE")
plt.legend();
plt.show()
```
