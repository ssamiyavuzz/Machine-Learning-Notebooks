## Lojistik Regresyon'un Varsaymlar (Assumptions)

**Sonu deikeni (label, class, target) kategoriktir**: Bu varsayma gre, tahmin edilmek istenen sonu deikeni kategoriktir, yani iki veya daha fazla kategoriye ait olabilir. rnein, geti/kald gibi iki snfl bir sonu veya ok snfl bir sonu olabilir.

**Gzlemler (veriler) birbirinden bamszdr**: Her gzlem, dier gzlemlerden bamsz olarak elde edilmi olmaldr. Bu, herhangi bir gzlemi dier gzlemlerin sonular zerinde etkisi olmadan gzlemlememiz gerektii anlamna gelir. Yani multicollinearity istemeyiz.

**Girdiler (deikenler) arasnda korelasyon bulunmaz**: Modelin doru bir ekilde almas iin girdi deikenleri (predictors) arasnda yksek bir korelasyon olmamaldr. Yksek korelasyonlu deikenler, modelin performansn azaltabilir ve tahminlerde yanltc sonulara neden olabilir.

**Lineer bir iliki ile z deeri hesapland iin x ile z arasnda lineer bir iliki varsayar**: Logistic regresyon modeli, girdi deikenleri ile kt deikeni arasnda lineer bir iliki varsayar. Yani, log-odds deerini hesaplamak iin kullanlan
$z$ deeri, girdi deikenlerinin lineer kombinasyonu olarak ifade edilir.

**Logistic regresyonun baarl olmas iin yksek miktarda veri lazmdr**: Logistic regresyon modelleri genellikle byk miktarda veriyle daha iyi performans gsterir. Bu, modelin doru bir ekilde renmesi ve genelleme yapabilmesi iin nemlidir. Ayrca, logistic regresyon modeli, lineer bir model olduu iin az miktarda veriyle kullanldnda ar uydurma (overfitting) eilimindedir ve yanltc sonular verebilir. Bu nedenle, modelin doru bir ekilde alabilmesi iin yeterli miktarda veriye ihtiya duyulur.

```python
from scipy.optimize import curve_fit

def sigmoid(x, L ,x0, k, b):
    y = L / (1 + np.exp(-k*(x-x0))) + b
    return y

# Scatter plot olutur
sns.scatterplot(x="physical_score", y="test_result", data=df)

# Sigmoid fonksiyonunu veriye uydur
popt, pcov = curve_fit(sigmoid, df['physical_score'], df['test_result'], method='dogbox', bounds=([0.,0.,0.,0.], [np.inf, np.inf, 1., np.inf]))

# Uygun sigmoid izgiyi iz
x_values = np.linspace(df['physical_score'].min(), df['physical_score'].max(), 200)
y_values = sigmoid(x_values, *popt)

plt.plot(x_values, y_values, color="red")

# Grafii gster
plt.show()
```

```python
from scipy.optimize import curve_fit

df = pd.read_csv('hearing_test.csv')

physical_score_min = df['physical_score'].min()
physical_score_max = df['physical_score'].max()
df['physical_score_normalized'] = (df['physical_score'] - physical_score_min) / (physical_score_max - physical_score_min)

def sigmoid(x, a, b):
    return 1 / (1 + np.exp(-(a*x + b)))

popt, _ = curve_fit(sigmoid, df['physical_score_normalized'], df['test_result'])

x_values = np.linspace(df['physical_score_normalized'].min(), df['physical_score_normalized'].max(), 100)
y_values = sigmoid(x_values, *popt)

sns.scatterplot(x='physical_score_normalized', y='test_result', data=df)

plt.plot(x_values, y_values, color='red', label='Sigmoid')

plt.xlabel('Normalized Physical Score')
plt.ylabel('Test Result')
plt.legend()
plt.show()
```

# Model Performance on Classification Tasks

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, log_loss
```

### Accuracy (Doruluk)

- Bir modelin tm tahminlerinin ne kadarnn doru olduunu gsterir.

- Doru tahminlerin toplam tahmin saysna orandr.

- Genel performans lsdr ama dengesiz snflarda yanltc olabilir.

```python
accuracy_score(y_test, y_pred)
```

```python
log_model.score(X_test_scaled, y_test) ## Accuracy ile ayn skoru retir = log model score
```

### Precision (Hassasiyet)

- Pozitif olarak tahmin edilen durumlarn gerekte ne kadarnn pozitif olduunu gsterir.

- Yanl pozitiflerin (false positives) etkisini minimize etmeye odaklanr.

- zellikle yanl pozitiflerin maliyetli olduu durumlarda nemlidir.

```python
precision_score(y_test, y_pred)
```

### Recall Score (Duyarllk)

- Gerekte pozitif olan durumlarn, model tarafndan ne kadarnn doru tahmin edildiini gsterir.

- Yanl negatifleri (false negatives) minimize etmeye odaklanr.

- zellikle yanl negatiflerin maliyetli olduu durumlarda nemlidir.

```python
recall_score(y_test, y_pred)
```

### f1_score

- Precision ve recall'un harmonik ortalamasdr.

- Dengeli bir lm salar, her iki metrii de dikkate alr.

- Hem yanl pozitifleri hem de yanl negatifleri nemli olduunda kullanlr.

### Log Los (Log Kayb)

- Tahmin edilen olaslklarn gerek deerlerden ne kadar sapma gsterdiinin bir lsdr.
- Modelin ne kadar "emin" olduunu ve bu eminliin gerekle ne kadar uyumlu olduunu deerlendirir.
- Daha hassas bir performans deerlendirmesi salar, zellikle olaslk temelli modellerde kullanlr.
