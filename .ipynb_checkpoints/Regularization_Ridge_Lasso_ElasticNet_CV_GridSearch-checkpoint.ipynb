```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (10,6)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("Advertising.csv")
df.head()
```

```python
X = df.drop("sales", axis=1)
y = df["sales"]
```

# Polynomial Conversion

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_converter = PolynomialFeatures(degree=5, include_bias=False)

poly_features = polynomial_converter.fit_transform(X)
```

```python
poly_features.shape
```

# Tran | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

# Data Scaling

**Scaling işlemi**, 

- Farklı feature'ların farklı aralıklarda olması durumunda, bu özelliklerin birbirleriyle daha uyumlu hale getirilmesi için yapılan bir ön işlemedir.

- Feature'ların ölçeklendirilmesi, algoritmanın daha hızlı ve daha etkin bir şekilde eğitilmesine yardımcı olur.

**Data Dengesizligi** -> elmadan 100 çilekten 500 olması

**Scaling** -> elmadaki ve çilekteki değerlerin sayısal olarak benzetilmeye çalışılması (belli bir aralığa ölçeklendirilmesi)

**-----------------------------------------------------------------------------**

**Standardization:**
- Feature'ların mean'ini 0 ve standart sapmasını 1 olacak şekilde ölçeklendirilmesi işlemidir. Bu işlem, feature'ların değerlerini ve dağılımını bir standard normal dağılıma (Gaussian dağılım) yakın hale getirir.
- **Bu yöntem, özellikle feature'lar arasındaki ilişkilerin korunması ve outlier'ların etkisinin azaltılması açısından faydalıdır.**

**Normalization:**
-  Feature'ların değerlerini belirli bir aralığa (genellikle 0 ile 1 arasında) ölçeklendirir. Bu işlem, özellikle min-max scaling yöntemiyle gerçekleştirilir.

- **Tüm feature'ların aynı ölçeğe sahip olmasını sağlar**, bu da bazı algoritmaların daha iyi performans göstermesine yardımcı olabilir.


**Robust Scaler:**
- Machine learning'te feature scaling yöntemlerinden biridir.

- Özellikle veri setindeki outliers'lardan etkilenmemek için tasarlanmıştır.


```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_train_scaled

# Train datasını belirtli bir ölçeğe oluştur
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled

# Test datasını data leakage(sızıntı) olmaması açısından train datasından ayrı transform ettik
```

```python
pd.DataFrame(X_train_scaled).agg(["mean", "std"]).round()
```

```python
pd.DataFrame(X_test_scaled).agg(["mean", "std"]).round()

# Yalnızca train datasında fit işlemi yaptığımız için test datasının min-max değeri farklı
# Bu yüzden test datasının std'sinde 0 ve 2'ler var
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train_scaled, y_train)
```

```python
y_pred = lm.predict(X_test_scaled)
y_train_pred = lm.predict(X_train_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, i):
    
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
              
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
ls = train_val(y_train, y_train_pred, y_test, y_pred, "linear")
ls
```

## Multicollinearity

***Multicollinearity***, bağımsız değişkenler arasında yüksek bir korelasyon olması durumunu ifade eder. Bir regresyon modelindeki iki veya daha fazla bağımsız değişkenin birbiriyle yüksek korelasyona sahip olması durumunda ortaya çıkar.
- Regresyon analizi gibi istatistiksel modellerde, bağımsız değişkenlerin (***Independent Variables***) ***target*** değişkeni (***Dependent Variables***) nasıl etkilediğini anlamaya çalışırız.

- Multicollinearity, bu bağımsız değişkenlerden ikisi veya daha fazlası arasında güçlü bir ilişki olduğunda ortaya çıkar. Bu durum, modelin tahmin gücünü etkileyebilir ve değişkenlerin katsayıları üzerindeki etkisini doğru bir şekilde yorumlamayı zorlaştırabilir.

- Featureların korelasyonunun **kendi arasında düşük**, **target ile** ise **çok yüksek** olmasını isteriz.

- Multicollinearity durumunu istemeyiz, overfitting'e neden olur.

- ***Örneğin:*** C ve F gibi, ikisi de aynı şeyleri söylüyor. İkisinin arasında ilişki çok yüksek.

```python
def color_red(val):
    if val > 0.90 and val < 0.99:
        color = 'red'
    elif val>=1:
        color = "blue"
    else:
        color = 'white'
    return f'color: {color}'

# Multicollinearity olan featurelar kırmızı olarak göster
```

```python
pd.DataFrame(poly_features).corr().style.applymap(color_red)
```

# Cross Validation (CV)

**Cross Validation (CV)**:
- Makine öğrenimi modellerinin performansını değerlendirme yöntemlerinden biridir.
- Modelin daha önce görmediği verilere olan genelleme yeteneğini test etmek için kullanılır.
- Temel amacı, modelin eğitim sırasında elde ettiği performansın gerçek dünya verileri üzerindeki performansını daha doğru bir şekilde tahmin etmektir.
- Modelin veri setinin farklı bölümlerine olan duyarlılığını test etmeye ve overfitting (aşırı uyum) gibi problemleri tespit etmeye yardımcı olur.
- Bu yöntem sayesinde, modelin farklı veri setlerine ne kadar iyi genelleme yapabildiği daha iyi anlaşılabilir.
