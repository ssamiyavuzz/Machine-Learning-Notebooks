```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (10,6)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("Advertising.csv")
df.head()
```

```python
X = df.drop("sales", axis=1)
y = df["sales"]
```

# Polynomial Conversion

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_converter = PolynomialFeatures(degree=5, include_bias=False)

poly_features = polynomial_converter.fit_transform(X)
```

```python
poly_features.shape
```

# Tran | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

# Data Scaling

**Scaling ilemi**, 

- Farkl feature'larn farkl aralklarda olmas durumunda, bu zelliklerin birbirleriyle daha uyumlu hale getirilmesi iin yaplan bir n ilemedir.

- Feature'larn leklendirilmesi, algoritmann daha hzl ve daha etkin bir ekilde eitilmesine yardmc olur.

**Data Dengesizligi** -> elmadan 100 ilekten 500 olmas

**Scaling** -> elmadaki ve ilekteki deerlerin saysal olarak benzetilmeye allmas (belli bir arala leklendirilmesi)

**-----------------------------------------------------------------------------**

**Standardization:**
- Feature'larn mean'ini 0 ve standart sapmasn 1 olacak ekilde leklendirilmesi ilemidir. Bu ilem, feature'larn deerlerini ve dalmn bir standard normal dalma (Gaussian dalm) yakn hale getirir.
- **Bu yntem, zellikle feature'lar arasndaki ilikilerin korunmas ve outlier'larn etkisinin azaltlmas asndan faydaldr.**

**Normalization:**
-  Feature'larn deerlerini belirli bir arala (genellikle 0 ile 1 arasnda) leklendirir. Bu ilem, zellikle min-max scaling yntemiyle gerekletirilir.

- **Tm feature'larn ayn lee sahip olmasn salar**, bu da baz algoritmalarn daha iyi performans gstermesine yardmc olabilir.


**Robust Scaler:**
- Machine learning'te feature scaling yntemlerinden biridir.

- zellikle veri setindeki outliers'lardan etkilenmemek iin tasarlanmtr.


```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train_scaled, y_train)
```

```python
y_pred = lm.predict(X_test_scaled)
y_train_pred = lm.predict(X_train_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, i):
    
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
              
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
ls = train_val(y_train, y_train_pred, y_test, y_pred, "linear")
ls
```

## Multicollinearity

***Multicollinearity***, bamsz deikenler arasnda yksek bir korelasyon olmas durumunu ifade eder. Bir regresyon modelindeki iki veya daha fazla bamsz deikenin birbiriyle yksek korelasyona sahip olmas durumunda ortaya kar.
- Regresyon analizi gibi istatistiksel modellerde, bamsz deikenlerin (***Independent Variables***) ***target*** deikeni (***Dependent Variables***) nasl etkilediini anlamaya alrz.

- Multicollinearity, bu bamsz deikenlerden ikisi veya daha fazlas arasnda gl bir iliki olduunda ortaya kar. Bu durum, modelin tahmin gcn etkileyebilir ve deikenlerin katsaylar zerindeki etkisini doru bir ekilde yorumlamay zorlatrabilir.

- Featurelarn korelasyonunun **kendi arasnda dk**, **target ile** ise **ok yksek** olmasn isteriz.

- Multicollinearity durumunu istemeyiz, overfitting'e neden olur.

- ***rnein:*** C ve F gibi, ikisi de ayn eyleri sylyor. kisinin arasnda iliki ok yksek.

```python
def color_red(val):
    if val > 0.90 and val < 0.99:
        color = 'red'
    elif val>=1:
        color = "blue"
    else:
        color = 'white'
    return f'color: {color}'

# Multicollinearity olan featurelar krmz olarak gster
```

```python
pd.DataFrame(poly_features).corr().style.applymap(color_red)
```

# Cross Validation (CV)

**Cross Validation (CV)**:
- Makine renimi modellerinin performansn deerlendirme yntemlerinden biridir.
- Modelin daha nce grmedii verilere olan genelleme yeteneini test etmek iin kullanlr.
- Temel amac, modelin eitim srasnda elde ettii performansn gerek dnya verileri zerindeki performansn daha doru bir ekilde tahmin etmektir.
- Modelin veri setinin farkl blmlerine olan duyarlln test etmeye ve overfitting (ar uyum) gibi problemleri tespit etmeye yardmc olur.
- Bu yntem sayesinde, modelin farkl veri setlerine ne kadar iyi genelleme yapabildii daha iyi anlalabilir.
